```python
# Q17. Write a Python program to calculate the roots of a quadratic equation.
import cmath  # Import the complex math module

def quadratic_roots(a, b, c):
    # Calculate the discriminant
    discriminant = b**2 - 4*a*c

    # Calculate the roots using the quadratic formula
    root1 = (-b + cmath.sqrt(discriminant)) / (2*a)
    root2 = (-b - cmath.sqrt(discriminant)) / (2*a)

    return root1, root2

# Get user input for coefficients
try:
    a = float(input("Enter coefficient a: "))
    b = float(input("Enter coefficient b: "))
    c = float(input("Enter coefficient c: "))

    roots = quadratic_roots(a, b, c)
    print(f"Root 1: {roots[0]}, Root 2: {roots[1]}")
except ValueError:
    print("Invalid input. Please enter valid numerical coefficients.")


```

```python
# Q20. Develop a program that checks if a given number is prime or not using nested if-else statements.

def is_prime(number):
    # Check if the number is less than 2 (prime numbers start from 2)
    if number < 2:
        return False
    # Check if the number is exactly 2 (the only even prime number)
    elif number == 2:
        return True
    # Check if the number is divisible by any integer from 2 to the square root of the number
    else:
        for i in range(2, int(number**0.5) + 1):
            if number % i == 0:
                return False
        return True

# Get user input for the number
try:
    num = int(input("Enter a number: "))
    if is_prime(num):
        print(f"{num} is a prime number.")
    else:
        print(f"{num} is not a prime number.")
except ValueError:
    print("Invalid input. Please enter a valid integer.")
```

```python
# Q27. Develop a program to calculate the electricity bill based on different consumption slabs using elif statements.
def calculate_electricity_bill(units_consumed):
    if units_consumed <= 50:
        bill_amount = units_consumed * 2.5
    elif 50 < units_consumed <= 100:
        bill_amount = 50 * 2.5 + (units_consumed - 50) * 3.0
    elif 100 < units_consumed <= 200:
        bill_amount = 50 * 2.5 + 50 * 3.0 + (units_consumed - 100) * 4.0
    elif 200 < units_consumed <= 300:
        bill_amount = 50 * 2.5 + 50 * 3.0 + 100 * 4.0 + (units_consumed - 200) * 5.0
    else:
        bill_amount = 50 * 2.5 + 50 * 3.0 + 100 * 4.0 + 100 * 5.0 + (units_consumed - 300) * 6.0

    return bill_amount

# Get user input for consumed units
try:
    units_consumed = float(input("Enter the number of units consumed: "))
    if units_consumed < 0:
        raise ValueError("Consumed units cannot be negative.")
except ValueError as e:
    print(f"Error: {e}")
else:
    # Calculate and display the electricity bill
    bill_amount = calculate_electricity_bill(units_consumed)
    print(f"The electricity bill for {units_consumed} units is: ${bill_amount:.2f}")
```

```python
# Q 28. Create a program to determine the type of quadrilateral based on its angles and sides using elif statements.
def determine_quadrilateral_type(angle1, angle2, angle3, angle4, side1, side2, side3, side4):
    # Check if it's a parallelogram
    if angle1 == angle3 and angle2 == angle4:
        if side1 == side3 and side2 == side4:
            print("It's a rectangle.")
        elif side1 == side2 == side3 == side4:
            print("It's a square.")
        else:
            print("It's a parallelogram.")
    # Check if it's a trapezoid
    elif angle1 == angle2 and angle3 != angle4:
        print("It's a trapezoid.")
    # Check if it's a kite
    elif side1 == side2 and side3 == side4:
        print("It's a kite.")
    # Check if it's a rhombus
    elif side1 == side2 == side3 == side4:
        print("It's a rhombus.")
    # If none of the above conditions are met
    else:
        print("It's a general quadrilateral.")

# Get user input for angles and sides
try:
    angle1 = float(input("Enter angle 1 in degrees: "))
    angle2 = float(input("Enter angle 2 in degrees: "))
    angle3 = float(input("Enter angle 3 in degrees: "))
    angle4 = float(input("Enter angle 4 in degrees: "))
    side1 = float(input("Enter side 1 length: "))
    side2 = float(input("Enter side 2 length: "))
    side3 = float(input("Enter side 3 length: "))
    side4 = float(input("Enter side 4 length: "))
except ValueError as e:
    print(f"Error: {e}")
else:
    # Determine and display the type of quadrilateral
    determine_quadrilateral_type(angle1, angle2, angle3, angle4, side1, side2, side3, side4)
```

```python
# Q15.Create a program that simulates a basic ATM transaction menu.
class ATM:
    def __init__(self, balance=1000):
        self.balance = balance

    def display_menu(self):
        print("\n===== ATM Transaction Menu =====")
        print("1. Check Balance")
        print("2. Withdraw Cash")
        print("3. Deposit Cash")
        print("4. Exit")

    def check_balance(self):
        print(f"\nYour balance is ₹{self.balance}")

    def withdraw_cash(self):
        amount = float(input("\nEnter the amount to withdraw: ₹"))
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"\nYou have successfully withdrawn ₹{amount}")
        else:
            print("\nInsufficient funds or invalid amount")

    def deposit_cash(self):
        amount = float(input("\nEnter the amount to deposit: ₹"))
        if amount > 0:
            self.balance += amount
            print(f"\nYou have successfully deposited ₹{amount}")
        else:
            print("\nInvalid amount")

    def run(self):
        self.display_menu()
        choice = input("\nEnter your choice (1-4): ")
        if choice == '1':
            self.check_balance()
        elif choice == '2':
            self.withdraw_cash()
        elif choice == '3':
            self.deposit_cash()
        elif choice == '4':
            print("\nThank you for using the ATM. Goodbye!")
        else:
            print("\nInvalid choice. Please enter a number between 1 and 4.")


atm = ATM(5000)
atm.run()
```

```python
#Q21. Create a program that calculates the income tax based on the user's income and tax brackets.
def calculate_income_tax(income):
    # Define tax brackets and their corresponding rates
    brackets = [(0, 10000, 0.1), (10001, 50000, 0.2), (50001, 100000, 0.3), (100001, float('inf'), 0.4)]

    # Calculate income tax
    tax = 0
    remaining_income = income

    for bracket in brackets:
        lower_limit, upper_limit, rate = bracket
        if remaining_income <= 0:
            break

        taxable_amount = min(upper_limit - lower_limit, remaining_income)
        tax += taxable_amount * rate
        remaining_income -= taxable_amount

    return tax

def main():
    # Get user input for income
    try:
        income = float(input("Enter your annual income: ₹"))
    except ValueError:
        print("Invalid input. Please enter a numerical value for income.")
        return

    # Calculate income tax
    income_tax = calculate_income_tax(income)

    # Display the result
    print(f"Your income tax for the year is: ₹{income_tax:.2f}")

main()
```

```python
#Q26. Write a program that simulates a basic quiz game with multiple-choice questions and scoring.

class Quiz:
    def __init__(self, questions):
        self.questions = questions
        self.score = 0

    def display_question(self, question):
        print(question['text'])
        for i, option in enumerate(question['options'], 1):
            print(f"{i}. {option}")
        return int(input("Enter your choice (1, 2, 3, or 4): "))

    def play(self):
        for question in self.questions:
            user_answer = self.display_question(question)
            correct_answer = question['correct_option']

            if user_answer == correct_answer:
                print("Correct!\n")
                self.score += 1
            else:
                print(f"Incorrect. The correct answer was option {correct_answer}.\n")

        print(f"Quiz completed! Your score: {self.score}/{len(self.questions)}")


def main():
    # Define quiz questions
    questions = [
        {
            'text': 'What is the capital of France?',
            'options': ['Berlin', 'Madrid', 'Paris', 'Rome'],
            'correct_option': 3
        },
        {
            'text': 'Which planet is known as the Red Planet?',
            'options': ['Mars', 'Jupiter', 'Venus', 'Saturn'],
            'correct_option': 1
        },
        {
            'text': 'Who wrote "Romeo and Juliet"?',
            'options': ['Charles Dickens', 'William Shakespeare', 'Jane Austen', 'Mark Twain'],
            'correct_option': 2
        }
    ]

    # Create a quiz object
    quiz = Quiz(questions)

    # Start the quiz
    quiz.play()

main()
```

```python
#Q29 .Build a program that determines the roots of a quartic equation using numerical methods.
def quartic_equation(x, coefficients):
    return sum(c * x**i for i, c in enumerate(coefficients))

def derivative_quartic(x, coefficients):
    return sum(i * c * x**(i - 1) for i, c in enumerate(coefficients) if i > 0)

def newton_raphson_quartic(coefficients, initial_guess, tolerance=1e-6, max_iterations=100):
    x = initial_guess

    for _ in range(max_iterations):
        f_x = quartic_equation(x, coefficients)
        f_prime_x = derivative_quartic(x, coefficients)

        if abs(f_prime_x) < tolerance:
            print("Derivative is too small. Newton-Raphson method may not converge.")
            return None

        x = x - f_x / f_prime_x

        if abs(f_x) < tolerance:
            return x

    print("Newton-Raphson method did not converge within the maximum number of iterations.")
    return None

def main():
    print("Quartic Equation Solver using Newton-Raphson Method")

    try:
        coefficients = [float(input(f"Enter coefficient for x^{i}: ")) for i in range(4, -1, -1)]
        initial_guess = float(input("Enter an initial guess for a root: "))
    except ValueError:
        print("Invalid input. Please enter numerical values.")
        return

    root = newton_raphson_quartic(coefficients, initial_guess)

    if root is not None:
        print(f"A root of the quartic equation is approximately: {root:.6f}")

if __name__ == "__main__":
    main()
```

```python
#Q31. Create a program that validates a password based on complexity rules (length, characters, etc.).

def is_valid_password(password):
    # Minimum length of the password
    min_length = 8

    # Check if the password meets the minimum length requirement
    if len(password) < min_length:
        return False

    # Check for at least one uppercase letter
    if not any(char.isupper() for char in password):
        return False

    # Check for at least one lowercase letter
    if not any(char.islower() for char in password):
        return False

    # Check for at least one digit
    if not any(char.isdigit() for char in password):
        return False

    # Check for at least one special character
    special_characters = "!@#$%^&*()-_=+[]{}|;:'\",.<>/?"
    if not any(char in special_characters for char in password):
        return False

    # If all checks pass, the password is valid
    return True

# Example usage:
password = input("Enter your password: ")
if is_valid_password(password):
    print("Password is valid!")
else:
    print("Password is not valid. Please ensure it meets the complexity requirements.")
```

```python
#Q35. Create a program that simulates a basic text-based tic-tac-toe game against the computer.
import random

def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board, player):
    # Check rows and columns
    for i in range(3):
        if all(board[i][j] == player for j in range(3)) or all(board[j][i] == player for j in range(3)):
            return True

    # Check diagonals
    if all(board[i][i] == player for i in range(3)) or all(board[i][2 - i] == player for i in range(3)):
        return True

    return False

def is_board_full(board):
    return all(board[i][j] != ' ' for i in range(3) for j in range(3))

def player_move(board):
    while True:
        try:
            row = int(input("Enter the row (1, 2, or 3): ")) - 1
            col = int(input("Enter the column (1, 2, or 3): ")) - 1

            if 0 <= row < 3 and 0 <= col < 3 and board[row][col] == ' ':
                return row, col
            else:
                print("Invalid move. Try again.")

        except ValueError:
            print("Invalid input. Please enter a number.")

def computer_move(board):
    available_moves = [(i, j) for i in range(3) for j in range(3) if board[i][j] == ' ']
    return random.choice(available_moves)

def play_tic_tac_toe():
    board = [[' ' for _ in range(3)] for _ in range(3)]
    current_player = 'X'

    while True:
        print_board(board)

        if current_player == 'X':
            row, col = player_move(board)
        else:
            print("Computer's move:")
            row, col = computer_move(board)

        board[row][col] = current_player

        if check_winner(board, current_player):
            print_board(board)
            print(f"{current_player} wins!")
            break
        elif is_board_full(board):
            print_board(board)
            print("It's a tie!")
            break

        current_player = 'O' if current_player == 'X' else 'X'


play_tic_tac_toe()
```

```python
#Q37. Develop a program that calculates the nth term of the Fibonacci sequence using memoization.
def fibonacci_with_memoization(n, memo={}):
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    result = fibonacci_with_memoization(n - 1, memo) + fibonacci_with_memoization(n - 2, memo)
    memo[n] = result
    return result

# Example usage:
term_number = int(input("Enter the term number in the Fibonacci sequence: "))

if term_number < 0:
    print("Please enter a non-negative number.")
else:
    fibonacci_result = fibonacci_with_memoization(term_number)
    print(f"The {term_number}th term in the Fibonacci sequence is: {fibonacci_result}")
```

```python
# Q38.Create a program that generates a calendar for a given month and year using conditional statements.
import calendar

def generate_calendar(year, month):
    # Check if the provided month and year are valid
    if not (1 <= month <= 12):
        print("Invalid month. Please enter a month between 1 and 12.")
        return

    # Use the calendar module to get the month's calendar
    cal = calendar.monthcalendar(year, month)

    # Print the calendar header
    print(f"\nCalendar for {calendar.month_name[month]} {year}")
    print("Mo Tu We Th Fr Sa Su")

    # Print each week in the calendar
    for week in cal:
        for day in week:
            if day == 0:
                print("   ", end="")
            else:
                print(f"{day:2} ", end="")
        print()

try:
    year = int(input("Enter the year: "))
    month = int(input("Enter the month (1-12): "))

    generate_calendar(year, month)
except ValueError:
    print("Invalid input. Please enter a valid year and month.")
```

```python
#Q39. Build a program that simulates a basic text-based blackjack game against the computer.
import random

def deal_card():
    """Return a random card from the deck."""
    cards = [11, 2, 3, 4, 5, 6, 7, 8, 9, 10, 10, 10, 10]
    return random.choice(cards)

def calculate_score(cards):
    """Calculate the total score of a hand."""
    # Check for Blackjack (Ace and a 10-valued card)
    if sum(cards) == 21 and len(cards) == 2:
        return 0

    # Convert Ace from 11 to 1 if the total score is over 21
    if 11 in cards and sum(cards) > 21:
        cards.remove(11)
        cards.append(1)

    return sum(cards)

def compare(player_score, computer_score):
    """Compare the player's and computer's scores to determine the winner."""
    if player_score == computer_score:
        return "It's a draw!"
    elif computer_score == 0:
        return "You lose! Computer has a Blackjack."
    elif player_score == 0:
        return "You win! You have a Blackjack."
    elif player_score > 21:
        return "You went over. You lose!"
    elif computer_score > 21:
        return "Computer went over. You win!"
    elif player_score > computer_score:
        return "You win!"
    else:
        return "You lose."

def play_game():
    print("Welcome to Blackjack!")

    player_cards = []
    computer_cards = []
    game_over = False

    # Deal initial cards
    for _ in range(2):
        player_cards.append(deal_card())
        computer_cards.append(deal_card())

    while not game_over:
        player_score = calculate_score(player_cards)
        computer_score = calculate_score(computer_cards)

        print(f"Your cards: {player_cards}, current score: {player_score}")
        print(f"Computer's first card: {computer_cards[0]}")

        # Check for Blackjack or bust
        if player_score == 0 or computer_score == 0 or player_score > 21:
            game_over = True
        else:
            # Ask the player if they want to draw another card
            should_continue = input("Type 'y' to get another card, 'n' to pass: ").lower()
            if should_continue == 'y':
                player_cards.append(deal_card())
            else:
                game_over = True

    # Let the computer draw cards
    while computer_score != 0 and computer_score < 17:
        computer_cards.append(deal_card())
        computer_score = calculate_score(computer_cards)

    print(f"Your final hand: {player_cards}, final score: {player_score}")
    print(f"Computer's final hand: {computer_cards}, final score: {computer_score}")

    # Compare scores and determine the winner
    print(compare(player_score, computer_score))

# Run the game
play_game()
```

```python
# Q40. Write a program that generates the prime factors of a given number using trial division.
def generate_prime_factors(number):
    prime_factors = []
    divisor = 2

    while divisor <= number:
        if number % divisor == 0:
            prime_factors.append(divisor)
            number //= divisor
        else:
            divisor += 1

    return prime_factors

# Example usage:
try:
    input_number = int(input("Enter a number to find its prime factors: "))

    if input_number <= 1:
        print("Please enter a number greater than 1.")
    else:
        factors = generate_prime_factors(input_number)
        print(f"Prime factors of {input_number}: {factors}")

except ValueError:
    print("Invalid input. Please enter a valid number.")
```
